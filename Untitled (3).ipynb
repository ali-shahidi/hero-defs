```python
xeno=[]
ch = {}
def find_strongest (group):
    for i in range(len(group)):
        xeno.append(len(user[i]["mission"]))
        ch[str(len(user[i]["mission"]))]=user[i]["name"]
    senku = max(xeno)
    return ch[str(senku)]
```

```python
find_strongests = find_strongest(user)
find_strongests 
```

```python
def generate_report(group):
    for i in range(len(group)):
      if len(group[i]["mission"])>=3:
         report = f"{group[i]["name"]} has {group[i]["power"]} power and {len(group[i]["mission"])} successfull missions."
         print(report)
```

```python
generate_report=generate_report(user)
generate_report
```

```python
def create_chat_log(filename):
    with open(filename,"w") as file:
        file.write("AI Chat Log\n")
        file.write("-----------\n")
def add_message(filename, speaker, message):
    with open(filename,"a") as file:
        file.write(f"{speaker}:{message}\n")
def show_chat_log(filename):
    with open(filename,"r") as file:
        content=file.read()
    print(content)   
```

```python
chat_filename = "chat_log.txt"
create_chat_log(chat_filename)
add_message(chat_filename, "User", "Hello")
add_message(chat_filename, "Bot", "Hi, how can I help you?")
add_message(chat_filename, "User", "What is AI?")
add_message(chat_filename, "Bot", "AI means Artificial Intelligence.")
show_chat_log(chat_filename)
```
